```python
import pandas as pd
```

```python
terrorism = pd.read_csv("terrorism.csv", encoding = "ISO-8859-1", usecols=[0, 1, 2, 3, 8, 12, 13, 14, 26, 29, 82, 84, 98, 101])
terrorism.dropna(inplace = True)
```

```python
terrorism.shape
```

```python
terrorism.columns = ["tid", "year", "month", "day", "country", "city", "latitude", "longitude", "success", "attack_type", "weapon_type", "weapon_subtype", "nkill", "nwound"]
```

```python
terrorism.to_csv("cleaned-terrorism.csv")
```
